1. คำนวณค่า correlation ระหว่างคู่ตัวแปร audio features ต่างๆ (3 points)
2. เลือกคู่ตัวแปรที่มีค่า correlation สูงที่สุดมาสร้าง Scatter plot เพื่อดูความสัมพันธ์​ (2 points) 
3. ความสัมพันธ์ของตัวแปรต่างๆ ในข้อ 1) อาจส่งผลอย่างไรในการทำโมเดล Machine Learning และเราสามารถหลีกเลี่ยงหรือลดปัญหาที่อาจจะเกิดขึ้นได้อย่างไร จงอภิปราย (2 points) 

```python
full_df_3.head()
```

```python
# PROBLEM 4.1: YOUR CODE HERE
from scipy.stats import pearsonr

c = [
  'danceability',
  'energy',
  'key',
  'loudness',
  'mode',
  'speechiness',
  'acousticness',
  'instrumentalness',
  'liveness',
  'valence',
  'tempo',
  'duration_ms',
  'time_signature'
]

max_i = 0
max_j = 0
max_cor = 0
for i in range(0, len(c)):
  for j in range(i + 1, len(c)):
    corr, _ = pearsonr(full_df_3[c[i]], full_df_3[c[j]])
    print('Pearsons correlation between {} and {} is {}'.format(c[i], c[j], corr))
    if corr > max_cor:
      max_cor = corr
      max_i = i
      max_j = j

print('Max Correlation:')
print('Pearsons correlation between {} and {} is {}'.format(c[max_i], c[max_j], max_cor))
```

```python
# PROBLEM 4.2: YOUR CODE HERE
import seaborn as sns
sns.scatterplot(data=full_df_3, x='energy', y='loudness', hue='genre')
```

PROBLEM 4.3: YOUR ANSWER HERE

จากกราฟ Energy และ Loudness มีค่า Pearsons Correlations แสดงว่า 2 ค่านี้มีความสัมพันธ์กันสูง อย่างไรก็ตาม Target Class หรือ Genre ก็ค่อนข้างกระจัดกระจาย ไม่ได้เกาะกลุ่มกันระหว่าง Genre เดียวกัน ทำให้ไม่ควรใช้เพียงแค่ 2 feature นี้ในการแบ่งกลุ่มข้อมูล

### Problem 13 (11 points)

2. จาก Playlist ทั้งหมด เราจะ**เลือกแค่ 5 Playlist แรกที่มีคำว่า "ร็อค" อยู่ในชื่อ Playlist** เพื่อหลีกเลี่ยงเพลงสากล สร้าง list เพื่อเก็บลิงก์ของ Playlists ทั้ง 5 อัน (4 points)
3. เปิดลิงก์ของแต่ละ Playlist เข้าไปเก็บข้อมูล **ชื่อเพลงและชื่อศิลปิน** โดยรวบรวมข้อมูลข้อมูลทั้งหมดลงใน `DataFrame` (4 points)

```python
# PROBLEM 13.1: YOUR CODE HERE
# โหลดข้อมูลจาก[หน้าเว็บ](https://www.joox.com/th/browse/2437)และดึง HTML Element ของทุก Playlist ในหน้านี้ออกมา (3 points)

import requests
from bs4 import BeautifulSoup

URL = 'https://www.joox.com/th/browse/2437'
response = requests.get(URL) 
document = BeautifulSoup(response.text,'html.parser')
divs = document.find_all('div', attrs={'class': 'col col-6 col-md-2'})

results = [div.a for div in divs]
results
```

```python
# PROBLEM 13.2: YOUR CODE HERE
# จาก Playlist ทั้งหมด เราจะเลือกแค่ 5 Playlist แรกที่มีคำว่า "ร็อค" อยู่ในชื่อ Playlist เพื่อหลีกเลี่ยงเพลงสากล สร้าง list เพื่อเก็บลิงก์ของ Playlists ทั้ง 5 อัน (4 points)
results_2 = [a for a in results if "ร็อค" in a.get('title')]
results_3 = results_2[0:5]
results_3
```

```python
# PROBLEM 13.3: YOUR CODE HERE
# เปิดลิงก์ของแต่ละ Playlist เข้าไปเก็บข้อมูล **ชื่อเพลงและชื่อศิลปิน** โดยรวบรวมข้อมูลข้อมูลทั้งหมดลงใน `DataFrame` (4 points)
song_result = []

for a in results_3:
  url = 'https://www.joox.com{}'.format(a.get('href'))
  response = requests.get(url) 
  document = BeautifulSoup(response.text,'html.parser')
  songs = document.find_all('div', attrs={'class': 'SongListItem'})
  for song in songs:
    a_song = song.find_all('a')
    song_result.append({
        'playlist': a.get('title'),
        'song': a_song[0].get_text(),
        'artist': a_song[1].get_text()
    })
df_song = pd.DataFrame(song_result)
df_song.shape
```
