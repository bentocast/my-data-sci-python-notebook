```python
import requests
from bs4 import BeautifulSoup

URL = 'https://medium.com/skooldio/tagged/productivity/'
response = requests.get(URL) 
document = BeautifulSoup(response.text,'html.parser')

```

```python
# ‡∏ä‡∏∑‡πà‡∏≠‡∏ö‡∏ó‡∏Ñ‡∏ß‡∏≤‡∏° (title)
# ‡∏ä‡∏∑‡πà‡∏≠‡∏£‡∏≠‡∏á (subtitle)
# ‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏µ‡πà‡πÇ‡∏û‡∏™‡∏ï‡πå (createdAt)
# ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏ö‡∏ó‡∏Ñ‡∏ß‡∏≤‡∏° (readingTime)
# ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô Claps (totalClapCount)

result = {}

title = document.find('h3').get_text()
subtitle = document.find('h4').get_text()
time = document.find('time').get('datetime')
readingTime = document.find('span', attrs={'class': 'readingTime'}).get('title')
totalClapCount = document.find('button',class_ = 'button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents').get_text()

print(title)
print(subtitle)
print(time)
print(readingTime)
print(totalClapCount)

# <button class="button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents" data-action="show-recommends" data-action-value="ef2201b9ae3c">428</button>
```

```python
title = document.find_all('h3', text='‡∏Ñ‡∏∏‡∏ì‡πÄ‡∏õ‡πá‡∏ô‡∏ú‡∏π‡πâ‡∏ô‡∏≥‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á Trust ‡πÑ‡∏î‡πâ‡∏î‡∏µ‡πÅ‡∏Ñ‡πà‡πÑ‡∏´‡∏ô?')
title
```

```python
all = document.find_all('div',class_= 'postArticle postArticle--short js-postArticle js-trackPostPresentation js-trackPostScrolls')
result=[]
for i in all:
  record = {}
  title = i.find('h3').get_text()
  record['title'] = title
  if i.find('h4') == None:
    subtitle = 'NULL'
  else:
    subtitle = i.find('h4').get_text()
  record['subtitle'] = subtitle
  time = i.find('time').get('datetime')
  record['time'] = time
  readingTime = i.find('span',class_='readingTime').get('title')
  record['readingTime'] = readingTime
  totalClapCount = i.find('button',class_ = 'button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents').get_text()
  record['totalClapCount'] = totalClapCount
  result.append(record)
print(result)
```
