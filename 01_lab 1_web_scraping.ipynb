## 1. Scraping data from a static web
จงเขียนโค้ดเพื่อดึงข้อมูลบทความที่ในหัวข้อ Productivity จาก Medium ของ Skooldio (https://medium.com/skooldio/tagged/productivity) โดยข้อมูลที่สนใจ คือ

* ชื่อบทความ (title)
* ชื่อรอง (subtitle)
* เวลาที่โพสต์ (createdAt) 
* ความยาวบทความ (readingTime)
* จำนวน Claps (totalClapCount)

*Note:*
* บางบทความอาจจะไม่มี subtitle ให้ใส่เป็น `null` หรือเว้นว่างไว้ 
* บางบทความอาจจะมี title หรือ subtitle ที่ยาวเกินไป เก็บข้อมูลมาแค่ที่ Display อยู่บนหน้าเว็บได้เลย 
* เวลาที่โพสต์ สามารถดึงตามที่ Display บนหน้าเว็บได้หรือ หรือถ้าใครหาเจอจะใส่เวลาเป๊ะๆ ก็ได้ (เช่น `2019-04-13T07:27:19.111Z` สำหรับตัวอย่างข้างต้น)
* ความยาวบทความขอเป็นตัวเลข (ไม่ใช่ข้อความ)


โดย output ออกมาเป็น list ของ dictionary หน้าตาเหมือนในตัวอย่างด้านล่างนี้

<img src="https://drive.google.com/uc?id=1uxguUFAv2szz9mXod2ZdZbe8GcuJPkX8" width="500"/>

`[{
  'title': 'ออกแบบ Talk ให้สนุกและทรงพลัง สไตล์ TED', 
  'subtitle': 'เบื้องหลัง Talk ในงาน Beta Conference 2019', 
  'createdAt': 'Apr 13, 2019'
  'readingTime': 6
  'totalClapCount': 231 
}]`

```python
import requests
from bs4 import BeautifulSoup

URL = 'https://medium.com/skooldio/tagged/productivity/'
response = requests.get(URL) 
document = BeautifulSoup(response.text,'html.parser')

```

```python
# ชื่อบทความ (title)
# ชื่อรอง (subtitle)
# เวลาที่โพสต์ (createdAt)
# ความยาวบทความ (readingTime)
# จำนวน Claps (totalClapCount)

result = {}

title = document.find('h3').get_text()
subtitle = document.find('h4').get_text()
time = document.find('time').get('datetime')
readingTime = document.find('span', attrs={'class': 'readingTime'}).get('title')
totalClapCount = document.find('button',class_ = 'button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents').get_text()

print(title)
print(subtitle)
print(time)
print(readingTime)
print(totalClapCount)

# <button class="button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents" data-action="show-recommends" data-action-value="ef2201b9ae3c">428</button>
```

```python
title = document.find_all('h3', text='คุณเป็นผู้นำที่สร้าง Trust ได้ดีแค่ไหน?')
title
```

```python
all = document.find_all('div',class_= 'postArticle postArticle--short js-postArticle js-trackPostPresentation js-trackPostScrolls')
result=[]
for i in all:
  record = {}
  title = i.find('h3').get_text()
  record['title'] = title
  if i.find('h4') == None:
    subtitle = 'NULL'
  else:
    subtitle = i.find('h4').get_text()
  record['subtitle'] = subtitle
  time = i.find('time').get('datetime')
  record['time'] = time
  readingTime = i.find('span',class_='readingTime').get('title')
  record['readingTime'] = readingTime
  totalClapCount = i.find('button',class_ = 'button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents').get_text()
  record['totalClapCount'] = totalClapCount
  result.append(record)
print(result)
```

## 2. Scraping data from an API 
เนื่องจากบทความบางส่วนถูกนำมาแสดงผลทีหลัง แสดงว่าหลังจากที่หน้าแรกโหลดมาแล้ว จะต้องมีการส่ง Request ผ่าน API เพื่อไปขอข้อมูลเพิ่มเติมมาทีหลัง ให้นิสิตต่อยอดโค้ดจากในส่วนแรก เพื่อดึงบทความที่ในหัวข้อ Data ***ทั้งหมด*** (นั่นคือ รวมบทความที่ถูกนำมาแสดงผลตอนแรกด้วย) โดยให้ output ออกมาเป็น list ของ dictionary เหมือนเดิม


#### 2.1) Identify the API URL for requesting additional data

```python
# Your Code Here
import requests
import json

url = " https://medium.com/skooldio/load-more?sortBy=tagged&tagSlug=data&limit=25&to=1527048907619"
```

#### 2.2) Request data from API 🔥🔥🔥

Hint: 
* API URL ที่ได้มา ถ้าหากส่ง Request ไปโดยไม่มีการระบุ Header เพิ่มเติม จะได้ Response ที่ไม่ถูกต้องกลับมา (ทดสอบได้โดยการลองเอา​ API URL ไปเปิดใน Web Browser)​ เพื่อให้ได้ผลลัพธ์ที่ถูกต้อง จะต้องมีการระบุ Header `Accept: application/json` ไปใน Request ด้วย ซึ่งสามารถทำได้โดยใช้คำสั่งนี้ `requests.get(url, headers={'accept': 'application/json'})`
* ข้อมูลที่ได้กลับมาจาก API จะไม่ใช่ json มาตรฐานและไม่สามารถใช้ `response.json()` ในการประมวลตามปกติได้ โดยจะต้องมีการลบอักขระพิเศษ `])}while(1);</x>` ที่ถูกเพิ่มเติมมาหน้า json ด้วยเหตุผลทางความปลอดภัย (ใครสนใจอ่านเพิ่มเติมได้[ที่นี่](https://stackoverflow.com/questions/2669690/why-does-google-prepend-while1-to-their-json-responses#:~:text=By%20placing%20a%20while(1,parse%20the%20remainder%20as%20JSON.) และ [นี่](https://security.stackexchange.com/questions/110539/how-does-including-a-magic-prefix-to-a-json-response-work-to-prevent-xssi-attack)) ออกก่อน ถึงจะสามารถประมวลผลได้ และใช้คำสั่ง `json.loads()` ([doc](https://www.w3schools.com/python/python_json.asp)) เพื่อ Parse json เป็น dictionary
* [Optional] เวลาที่ได้จาก API จะเป็น Unix time สามารถแปลงให้เป็นข้อความในรูปแบบเดียวกับบทความก่อนหน้านี้ได้ สามารถดูตัวอย่างได้[ที่นี่](https://www.tutorialspoint.com/How-to-convert-unix-timestamp-string-to-readable-date-in-Python)

```python
# Your Code Here
import datetime

# timestamp = datetime.datetime.fromtimestamp(1500000000)
# print(timestamp.strftime('%Y-%m-%d %H:%M:%S'))

response = requests.get(url,headers = {'accept':'application/json'})
data = json.loads(response.text[16:])
post = data['payload']['references']['Post']
result = []
for value in post.values():
  record = {}
  result.append({
      'title':value['title'],
      'subtitle':value['previewContent']['subtitle'],
      'time':datetime.datetime.fromtimestamp(int(value['firstPublishedAt'])/1000.0).strftime('%Y-%m-%d %H:%M:%S'),
      'readingTime':value['virtuals']['readingTime'],
      'totalClapCount':value['virtuals']['totalClapCount']
      })
print(result)
```
