เพื่อทำความเข้าใจข้อมูลในตาราง transaction มากขึ้น จงคำนวณค่าดังต่อไปนี้ (รวม 6 ค่า)

```
print(trx['SHOP_DATE'].min())
print(trx['SHOP_DATE'].max())

print(trx['PROD_CODE'].nunique())
print(trx['PROD_CODE_30'].nunique())

print(trx['SHOP_HOUR'].min())
print(trx['SHOP_HOUR'].max())
```

(1.4) 
เพื่อทำความเข้าใจกลุ่มลูกค้ามากยิ่งขึ้น จากตาราง customers จงแสดงรายชื่อประเภท Lifestage พร้อมจำนวนลูกค้าในกลุ่มนั้น เรียงตามขนาดกลุ่มลูกค้าจากใหญ่ไปเล็ก นอกจากนี้ สำหรับกลุ่มลูกค้าที่ค่า Lifestage เป็น NULL ให้แสดงผลเป็นข้อความ ‘XX’ แทน 


```
df = customers.fillna('XX').groupby('CUST_LIFESTAGE')[['CUST_CODE']].count()
df.columns = ['COUNT']
print(df.sort_values(['COUNT'], ascending=False))
```

(1.5)
