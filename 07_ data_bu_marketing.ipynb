```python
from numpy import random
import math
import statistics

# def runSim(cr, ctr, c1, bid):
#     vol = math.sqrt(bid)
#     impr = random.binomial(n=(c1*vol), p=0.5, size=1)
#     clicks = random.binomial(n=impr, p = ctr, size = 1)
#     orders = random.binomial(n=clicks, p = cr, size = 1)
#     tot_orders = random.binomial(n=500, p = 0.5, size = 1) + random.binomial(n=orders, p=0.3, size = 1)
#     return impr, clicks, orders, tot_orders

def mkSim(numGames, cr, ctr1,ctr2, c1, bid, ppo):
    impr_t = []
    clicks_t = []
    orders_t = []
    nc_t = []
    trial_t = []
    impr_t_sq = []
    clicks_t_sq = []
    orders_t_sq = []
    nc_t_sq = []
    tot_orders_t = []
    tot_orders_t_sq = []
    tot_nc_t = []
    tot_nc_t_sq = []
    for t in range (numGames):
        impr, clicks, orders, tot_orders = runSim(cr, ctr1,ctr2, c1, bid)
        nc = orders*ppo - clicks*bid
        nc_sq = (orders*ppo - clicks*bid)**2
        tot_nc = tot_orders*ppo - clicks*bid
        tot_nc_sq = (tot_orders*ppo - clicks*bid)**2
        impr_sq = impr**2
        clicks_sq = clicks**2
        orders_sq = orders**2
        tot_orders_sq = tot_orders**2
        trial = t
        trial_t.append(trial)
        impr_t.append(impr)
        impr_t_sq.append(impr_sq)
        clicks_t.append(clicks)
        clicks_t_sq.append(clicks_sq)
        orders_t.append(orders)
        orders_t_sq.append(orders_sq)
        tot_orders_t.append(tot_orders)
        tot_orders_t_sq.append(tot_orders_sq)
        nc_t.append(nc)
        nc_t_sq.append(nc_sq)
        tot_nc_t.append(tot_nc)
        tot_nc_t_sq.append(tot_nc_sq)
    print("Number of trials: ", numGames)
    impr_mu = sum(impr_t)/numGames
    print("Mean impr: ", impr_mu)
    impr_sigma = math.sqrt((sum(impr_t_sq) - 2*impr_mu*sum(impr_t) + numGames*impr_mu**2)/(numGames - 1))
    print("Stdev impr: ", impr_sigma)
    impr_mu_se = impr_sigma / math.sqrt(numGames)
    print("SE impr: ", impr_mu_se)
    clicks_mu = sum(clicks_t)/numGames
    print("Mean clicks: ", clicks_mu)
    clicks_sigma = math.sqrt((sum(clicks_t_sq) - 2*clicks_mu*sum(clicks_t) + numGames*clicks_mu**2)/(numGames - 1))
    print("Stdev clicks: ", clicks_sigma)
    clicks_mu_se = clicks_sigma / math.sqrt(numGames)
    print("SE clicks: ", clicks_mu_se)
    orders_mu = sum(orders_t)/numGames
    print("Mean orders: ", orders_mu)
    orders_sigma = math.sqrt((sum(orders_t_sq) - 2*orders_mu*sum(orders_t) + numGames*orders_mu**2)/(numGames - 1))
    print("Stdev orders: ", orders_sigma)
    orders_mu_se = orders_sigma / math.sqrt(numGames)
    print("SE orders: ", orders_mu_se)
    tot_orders_mu = sum(tot_orders_t)/numGames
    print("Mean total orders: ", tot_orders_mu)
    tot_orders_sigma = math.sqrt((sum(tot_orders_t_sq) - 2*tot_orders_mu*sum(tot_orders_t) + numGames*tot_orders_mu**2)/(numGames - 1))
    print("Stdev total orders: ", tot_orders_sigma)
    tot_orders_mu_se = tot_orders_sigma / math.sqrt(numGames)
    print("SE total orders: ", tot_orders_mu_se)
    nc_mu = sum(nc_t)/numGames
    print("Mean channel net contribution: ", nc_mu)
    nc_sigma = math.sqrt((sum(nc_t_sq) - 2*nc_mu*sum(nc_t) + numGames*nc_mu**2)/(numGames - 1))
    print("Stdev channel NC: ", nc_sigma)
    nc_mu_se = nc_sigma / math.sqrt(numGames)
    print("SE channel NC: ", nc_mu_se)
    tot_nc_mu = sum(tot_nc_t)/numGames
    print("Mean total net contribution: ", tot_nc_mu)
    tot_nc_sigma = math.sqrt((sum(tot_nc_t_sq) - 2*tot_nc_mu*sum(tot_nc_t) + numGames*tot_nc_mu**2)/(numGames - 1))
    print("Stdev total NC: ", tot_nc_sigma)
    tot_nc_mu_se = tot_nc_sigma / math.sqrt(numGames)
    print("SE total NC: ", tot_nc_mu_se)
```

```python
def runSim(cr, ctr1,ctr2, c1, bid):
    vol = math.sqrt(bid)
    impr1 = random.binomial(n=(c1*vol), p=0.05, size=1)
    impr2 = random.binomial(n=(c1*vol), p=0.1, size=1)

    clicks1 = random.binomial(n=impr1, p = ctr1, size = 1)
    clicks1_1 = random.binomial(n=clicks1, p = ctr1, size = 1)

    clicks2 = random.binomial(n=impr2, p = ctr2, size = 1)
    clicks = clicks1_1+clicks2

    orders = random.binomial(n=clicks, p = cr, size = 1)
    tot_orders = random.binomial(n=500, p = 0.5, size = 1) + random.binomial(n=orders, p=0.3, size = 1)
    return impr1+impr2, clicks, orders, tot_orders
```

```python
impr, clicks, orders, tot_orders =  runSim(cr_tk, ctr_tk,ctr_go, c1_tk, bid_tk)
```
